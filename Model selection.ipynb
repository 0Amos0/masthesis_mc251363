```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
```

```python
data=pd.read_excel('C:/Users/Amos/Desktop/MC251361/Python/Ep_raw.xlsx')
```

```python
df=data
y = df['Ep'].copy()
X = df.drop('Ep', axis=1)
```

```python

# 
n_runs = 100  # 
test_size = 0.2  # 

# 
models = {
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Support Vector Machine": SVR(),
    "Neural Network": MLPRegressor(max_iter=1000, random_state=42)
}

# 
results = {model_name: {'R2': [], 'RMSE': []} for model_name in models.keys()}

# R2RMSE
for i in range(n_runs):
    # 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=i)
    print(f"\nRun {i + 1}/{n_runs}:")

    for model_name, model in models.items():
        # 
        model.fit(X_train, y_train)
        # 
        y_pred = model.predict(X_test)
        # R2RMSE
        r2 = r2_score(y_test, y_pred)
        rmse = sqrt(mean_squared_error(y_test, y_pred))
        
        # R2RMSE
        results[model_name]['R2'].append(r2)
        results[model_name]['RMSE'].append(rmse)
        
        # R2RMSE
        print(f"{model_name} R^2: {r2:.4f}")
        print(f"{model_name} RMSE: {rmse:.4f}")
```

```python
# 
summary = {}

for model_name, model_results in results.items():
    avg_r2 = np.mean(model_results['R2'])
    std_r2 = np.std(model_results['R2'])
    avg_rmse = np.mean(model_results['RMSE'])
    std_rmse = np.std(model_results['RMSE'])
    
    summary[model_name] = {'Average R^2': avg_r2, 'R^2 Std Dev': std_r2,
                           'Average RMSE': avg_rmse, 'RMSE Std Dev': std_rmse}

#  pandas DataFrame
summary_df = pd.DataFrame(summary).T  # '.T' 

# 
print(summary_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")  # 
palette = sns.color_palette("Set2", len(models))  # 

# 
ax = sns.boxplot(data=[results[model]['R2'] for model in models.keys()], palette=palette)

# 
sns.stripplot(data=[results[model]['R2'] for model in models.keys()], color='black', alpha=0.5, jitter=True)

# 
plt.ylabel('R^2 Score', fontsize=14)

# 
ax.tick_params(axis='both', labelsize=12)

# x
plt.xticks(ticks=range(len(models)), labels=models.keys(), fontsize=12)

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.datasets import make_regression
from xgboost import XGBRegressor
import lightgbm as lgb
from math import sqrt


# 
n_runs = 100  # 
test_size = 0.2  # 

# 
models = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42),
    "LightGBM": lgb.LGBMRegressor(random_state=42)
}

# 
results = {model_name: {'R2': [], 'RMSE': []} for model_name in models.keys()}

# R2
for i in range(n_runs):
    # 
    random_state = i
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    print(f"\nRun {i + 1}/{n_runs}: Random state = {random_state}")

    for model_name, model in models.items():
        # 
        model.fit(X_train, y_train)
        # 
        y_pred = model.predict(X_test)
        # R2RMSE
        r2 = r2_score(y_test, y_pred)
        rmse = sqrt(mean_squared_error(y_test, y_pred))
        
        # R2RMSE
        results[model_name]['R2'].append(r2)
        results[model_name]['RMSE'].append(rmse)
        
        # R2RMSE
        print(f"{model_name} R^2: {r2:.4f}")
        print(f"{model_name} RMSE: {rmse:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")  # 
palette = sns.color_palette("Set2", len(models))  # 

# 
ax = sns.boxplot(data=[results[model]['R2'] for model in models.keys()], palette=palette)

# 
sns.stripplot(data=[results[model]['R2'] for model in models.keys()], color='black', alpha=0.5, jitter=True)

# 
plt.ylabel('R^2 Score', fontsize=14)

# 
ax.tick_params(axis='both', labelsize=12)

# x
plt.xticks(ticks=range(len(models)), labels=models.keys(), fontsize=12)

# 
plt.show()
```
