```python
data2=pd.read_excel('C:/Users/Amos/Desktop/MC251361/Python/Ep_val.xlsx')
```

```python
df1=data1
y1 = df1['Ep'].copy()
X1= df1.drop('Ep', axis=1)
```

```python
df2=data2
y2 = df2['Ep'].copy()
X2 = df2.drop('Ep', axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from lightgbm import LGBMRegressor
from math import sqrt

# Define model parameters
param_space = {
    'n_estimators': 177,
    'max_depth': 5,
    'learning_rate': 0.07809765271680998,
    'num_leaves': 150,
    'min_child_samples': 5,
    'subsample': 0.6,
    'colsample_bytree': 0.6
}

# Initialize LightGBM model
LGBM = LGBMRegressor(**param_space, random_state=42)

# Train-test split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=63)

# Train the model
LGBM.fit(X_train1, y_train1)

# Make predictions
y_pred_test1 = LGBM.predict(X_test1)

# Calculate R2 and RMSE for train and test sets
r2_test1 = r2_score(y_test1, y_pred_test1)
rmse_test1 = sqrt(mean_squared_error(y_test1, y_pred_test1))

# Print results
print(f'Test R21: {r2_test1}')
print(f'Test RMSE1: {rmse_test1}')
```

```python
X2
```

```python
# Make predictions
y_pred_test2 = LGBM.predict(X2)

# Calculate R2 and RMSE for train and test sets
r2_test2 = r2_score(y2, y_pred_test2)
rmse_test2 = sqrt(mean_squared_error(y2, y_pred_test2))

# Print results
print(f'Test R22: {r2_test2}')
print(f'Test RMSE2: {rmse_test2}')
```

```python
#Test stage
plt.scatter(y2, y_pred_test2, alpha=0.7)
plt.plot([y2.min(), y2.max()], [y_pred_test2.min(), y_pred_test2.max()], color='red', linestyle='--')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Vale')
plt.text(0.05, 0.95, f'R²: {(r2_test2):.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
```
